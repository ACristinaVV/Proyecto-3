```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from ipywidgets import *
from datetime import date
```

```python
def interes_compuesto(C0,meta,i):
    #Despejamos k
    k=np.ceil(np.log(meta/C0)/np.log(1+i))  
    k=k.astype(int)                        
    Ck=C0*(1+i)**k                         
    Ck=round(Ck,2)                          
    
    #Gráfica
    periodos=np.linspace(0,k,k+1)   
    Capitales=C0*(1+i)**periodos    
    plt.figure(num=1)              
    plt.clf()                      
    plt.plot(periodos,Capitales,'*',ms=10)
    plt.xlabel('$k$')
    plt.ylabel('$C_k$')
    plt.plot([0,k],[meta,meta],'--k')  
    plt.grid()
    plt.show()
    print('El número de meses requeridos para alcanzar ',meta, ' pesos es: ',k,
      '. El capital generado es de ',Ck,' pesos.',sep='')
    
```

```python
interes_compuesto(600000,900000,.00805) #BBVA mensual
```

```python
interes_compuesto(600000,900000,.018 ) #Banorte trimestral
```

```python
interes_compuesto(600000,900000,.01382) #HSBC (bimestral)
```
